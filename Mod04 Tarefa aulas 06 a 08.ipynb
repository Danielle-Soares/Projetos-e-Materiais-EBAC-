````python
#1) Seu código aqui
def consulta_uf_municipio (uf=str,cidade=str):

    dados_uf=dados[dados['uf']==uf]
    resposta=dados_uf[dados_uf['interessado']==cidade]
    return resposta
````

````python
consulta_uf_municipio('PR','Foz do Iguaçu')
````

````python
# 2) Seu código aqui
consulta_MG=dados[dados['uf']=='MG']
consulta_MG=consulta_MG[consulta_MG['status']=='Arquivado por decurso de prazo']
consulta_MG
````
