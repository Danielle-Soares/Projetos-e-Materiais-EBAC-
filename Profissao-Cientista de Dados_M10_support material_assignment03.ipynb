````python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
````

````python
df=pd.read_csv('previsão_de_renda.csv')
df
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
tabela=pd.crosstab(df['data_ref'],df['sexo'])
grafico=sns.countplot(x='data_ref',data=df,hue='sexo')
plt.legend(loc='upper right')
tabela.plot.bar(stacked=True)
grafico.set_xticklabels(df['data_ref'].unique(),rotation=90)
plt.legend(bbox_to_anchor=(1,1))
````

````python
tabela1=pd.crosstab(df['data_ref'],df['tipo_renda'])
grafico=sns.countplot(x='data_ref',data=df,hue='tipo_renda')
plt.legend(loc='upper right')
tabela1.plot.bar(stacked=True)
grafico.set_xticklabels(df['data_ref'].unique(),rotation=90)
plt.legend(bbox_to_anchor=(1,1))
````

````python
tabela2=pd.crosstab(df['data_ref'],df['educacao'])
grafico=sns.countplot(x='data_ref',data=df,hue='educacao')
plt.legend(loc='upper right')
tabela2.plot.bar(stacked=True)
grafico.set_xticklabels(df['data_ref'].unique(),rotation=90)
plt.legend(bbox_to_anchor=(1,1))
grafico.figure.set_size_inches(12, 8)
````

````python
tabela3=pd.crosstab(df['data_ref'],df['estado_civil'])
grafico=sns.countplot(x='data_ref',data=df,hue='estado_civil')
tabela3.plot.bar(stacked=True)
grafico.set_xticklabels(df['data_ref'].unique(),rotation=90)
plt.legend(bbox_to_anchor=(1,1))
grafico.figure.set_size_inches(12, 8)
````

````python
tabela4=pd.crosstab(df['data_ref'],df['tipo_residencia'])
grafico=sns.countplot(x='data_ref',data=df,hue='tipo_residencia')
tabela4.plot.bar(stacked=True)
grafico.set_xticklabels(df['data_ref'].unique(),rotation=90)
plt.legend(bbox_to_anchor=(1,1))
grafico.figure.set_size_inches(12, 8)
````
