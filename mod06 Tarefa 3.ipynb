#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[df['APGAR5']<4,'categoria_APGAR5']='asfixia_grave'
df.loc[df['APGAR5']<6,'categoria_APGAR5']='asfixia_moderada'
df.loc[df['APGAR5']<8,'categoria_APGAR5']='asfixia_leve'
df.loc[df['APGAR5']>7,'categoria_APGAR5']='normal'
df[['APGAR5','categoria_APGAR5']].head(20)
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.catplot(y='APGAR5', x='GESTACAO',hue='SEXO', kind='bar', data=df).tick_params(axis="x", rotation=90)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabela=pd.crosstab(df['GESTACAO'],df['APGAR5'] )
tabela
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tabela.div(tabela.sum(axis=1),axis=0)
```

```python
tabela.plot.bar()
```

O gráfico corrobora com a afirmação dos médicos, pois podemos verificar que bebês nascidos com menos de 37 semanas apresentam APGAR baixo, demonstrando algum grau de asfixia.
