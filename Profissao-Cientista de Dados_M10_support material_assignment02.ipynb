1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
previsao_renda=pd.read_csv('previsão_de_renda.csv')
qualitativas=previsao_renda[['posse_de_veiculo','posse_de_imovel']]
qualitativas
````

````python
sns.pointplot(x='posse_de_veiculo', y=previsao_renda['renda'],data=qualitativas,hue='posse_de_imovel',dodge=True)
````

````python
sns.pointplot(x='posse_de_imovel', y=previsao_renda['renda'],data=qualitativas,hue='posse_de_veiculo',dodge=True)
````

````python
sns.pointplot(x=previsao_renda['educacao'], y=previsao_renda['renda'],data=qualitativas,hue='posse_de_imovel',dodge=True)
plt.xticks(rotation=45)
````

````python
sns.pointplot(x=previsao_renda['educacao'], y=previsao_renda['renda'],data=qualitativas,hue='posse_de_veiculo',dodge=True)
plt.xticks(rotation=45)
````

````python
sns.pointplot(x=previsao_renda['estado_civil'], y=previsao_renda['renda'],data=qualitativas,hue='posse_de_imovel',dodge=True)
````

````python
sns.pointplot(x=previsao_renda['estado_civil'], y=previsao_renda['renda'],data=qualitativas,hue='posse_de_veiculo',dodge=True)
````

````python
sns.pointplot(x=previsao_renda['tipo_residencia'], y=previsao_renda['renda'],data=qualitativas,hue='posse_de_imovel',dodge=True)
plt.xticks(rotation=45)
````

````python
sns.pointplot(x=previsao_renda['tipo_residencia'], y=previsao_renda['renda'],data=qualitativas,hue='posse_de_veiculo',dodge=True)
plt.xticks(rotation=45)
````

## A variável posse de veículo parece ter mais significância  que a variável posse de imóvel, quando comparadas as outras variáveis dos dados.
